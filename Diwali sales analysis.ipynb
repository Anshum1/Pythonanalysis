***GENDER***

```python
ax=sns.countplot(x='Gender',data=df) #count of data

for bars in ax.containers: #to show values in a container
    ax.bar_label(bars)
    
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(hue= 'Gender' ,y= 'Amount', data=sales_gen,palette='RdPu')

```

**From the graphs below we can see that most of the buyers are females and have more purchasing power than men!***

***AGE***

```python
ax=sns.countplot(data=df,x='Age Group', hue = 'Gender') #hue shows male and female graph

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(hue='Age Group', y='Amount',data=sales_age,palette='hls')
```

**From above graphs we can see that most of the Buyers are the age between 26-35 yrs.***


***STATE***

```python
df.columns
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10) #.head to show only 10 states
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state, hue='State',y='Orders',color='blue',palette='RdBu')
```

**From the above graphs, we analyse that most of orders and total sales are from UP,Maharastra and Karnataka in top 3 spots.***

***Marital Status***

```python
ax=sns.countplot(data=df,x='Marital_Status',color='Orange') #hue shows male and female graph

sns.set(rc={'figure.figsize':(5,7)})
ax.bar_label(ax.containers[0])
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,7)})
sns.barplot(data=sales_state, x='Marital_Status',y='Amount',color='blue',palette='hls',hue='Gender')
```

**From above we conclude that most of the buyers are married (women in this case) and they have high purchasing power!**

***Occupation***

```python
sns.set(rc={'figure.figsize':(20,7)})
ax=sns.countplot(data=df,x='Occupation',hue='Gender',palette='Paired') #hue shows male and female graph

ax.bar_label(ax.containers[0])
```

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state, hue='Occupation',y='Amount',palette='tab10') 
```

**From above graphs we find that most of the buyers are working in IT, Healthcare and Aviation sentor**

***Product Category***


```python
sns.set(rc={'figure.figsize':(25,7)})
ax=sns.countplot(x='Product_Category',data=df) 

ax.bar_label(ax.containers[0])
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state, hue='Product_Category',y='Amount',palette='husl')
```

**From the above we notice that most of the sold products are from Food, Footwear and Electronics category**
